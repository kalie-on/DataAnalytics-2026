```python
def favorite_things(name, movie, music):

    return f"""
Name: {name}
Favorite Movie: {movie}
Favorite Musician/Band: {music}
"""
```

```python
print(favorite_things(
    "Caleb",
    "Black Panther",
    "Drake"
))
```

```python
def why_im_here(purpose):
   return purpose
```

```python
print(why_im_here(
    "I joined Year Up United to gain skills in data analytics and start a strong career in technology."
))
```

```python
def favorite_place(place, why):

    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place(
    "atlanta",
    "it has amazing views and entertainment"
))
```

```python
print(favorite_place(
    "California",
    "the weather and beaches are beautiful"
))
```
