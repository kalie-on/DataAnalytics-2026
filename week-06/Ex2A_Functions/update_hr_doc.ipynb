## Lab 3. Employee data

This lab exercise makes updates to information extracted from
an employee database.

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
print(hr_list)
```

```python
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
```

```python
print(hr_list)
```

```python
for employee in hr_list:
    employee_num = employee[0]
    dept_code = employee[1]
    name = employee[2]
    salary = int(employee[3])

    print(f"{employee_num} | {dept_code} | {name} | {salary:,}")
```
