# Exercise 2.B Working with datetime functions

## Lab 2 - Times and Dates
This lab demonstrates how to work with dates, times, and datetime functions in Python.

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
print(today)
```

```python
today = today.strftime("%A, %B %d, %Y")
print(f"Today is {today}")
```

# My birthday

```python
birthday = datetime.date(2004, 2, 21)
print(birthday)
```

```python
my_birthday = birthday.strftime("%m/%d/%Y")
print(f"My birthday is {my_birthday}")
```

# Next 90 Days

```python
today = datetime.datetime.now()
ninety_d = today + datetime.timedelta(days=90)
print(ninety_d)
```

```python
formatted_ninety = ninety_d.strftime("%B %d, %Y")
print(f"90 days from today is {formatted_ninety}")
```

# Time for Dinner

```python
dinner_time = datetime.time(7 , 40)
print(dinner_time)
```

```python
formatted_dinner = dinner_time.strftime("%I:%M %p")
print(f"Let's meet for dinner at {formatted_dinner}")
```
