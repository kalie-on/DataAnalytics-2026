```python
def get_soc_sec_tax(gross_pay):

    return gross_pay * 0.062
```

```python
def get_medicare_tax(gross_pay):

    return gross_pay * 0.0145
```

```python
def get_federal_tax(gross_pay, withholding_code):

    if withholding_code == 0:
        rate = 0.23

    elif withholding_code == 1:
        rate = 0.21

    elif withholding_code == 2:
        rate = 0.195

    elif withholding_code == 3:
        rate = 0.185

    else:
        rate = 0.18 - ((withholding_code - 4) * 0.005)

    return gross_pay * rate
```

```python
print(get_soc_sec_tax(750))
print(get_medicare_tax(750))
print(get_federal_tax(750, 0))
```

```python
print(get_soc_sec_tax(1550))
print(get_medicare_tax(1550))
print(get_federal_tax(1550, 2))
```

```python
print(get_soc_sec_tax(1100))
print(get_medicare_tax(1100))
print(get_federal_tax(1100, 5))
```
